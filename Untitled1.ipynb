```python
import math
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from numpy import *
```

```python
fs = 2 * 10**3     ## 2kHz sampling rate
fx = [500, 900, 1.1 * 10**3, 1.5 * 10**3, 2.2 * 10**3]
t = arange(0, float(10) * 10**(-3), float(10) * 10**(-3) / 500)     ## Span of 5 ms
ts = arange(-15 * 10**(-3), 25 * 10**(-3), float(1)/fs)             ## Sampled time
```

```python
t
```

```python
def Contiguous_sinusoidal_signal(t, fx):
    return np.cos( 2 * np.pi * fx * t )
```

```python
for fxi in fx:
    plot(t, Contiguous_sinusoidal_signal(t, fxi))
    stem(ts, Contiguous_sinusoidal_signal(ts, fxi), linefmt='r-', markerfmt='ro', basefmt='k-')
    plt.show()
```
