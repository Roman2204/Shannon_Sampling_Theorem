```python
import math
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from numpy import *
```

```python
f = 1.0
fs = 5.0
Cycles_of_sinusoid = 5
time = linspace(0, float(Cycles_of_sinusoid)/f, 1000)
#time_sampling = linspace(0, float(Cycles_of_sinusoid)/f, (float(Cycles_of_sinusoid)/f) / ( float(1)/(2*fs) ) + 1 )
time_sampling = linspace(-float(Cycles_of_sinusoid)/f,
                         2*float(Cycles_of_sinusoid)/f,
                         3 * ( (float(Cycles_of_sinusoid)/f) / ( float(1)/(2*fs) ) + 1 ) )
```

```python
time_sampling
```

```python
def Contiguous_sinusoidal_signal(t):
    return np.cos( 2 * np.pi * f * t )
```

```python
def Shannon_Function(t, fs):   
    sum = 0
    for n in range( len(time_sampling) ):
        sum += Contiguous_sinusoidal_signal(time_sampling[n]) * np.sin( 2*fs*t*math.pi - n*np.pi ) / ( 2*fs*t*np.pi - n*np.pi )
    return sum
```

```python
def graph_signal(ax):
    
    ax.plot(time, Contiguous_sinusoidal_signal(time))
    ax.stem(time_sampling, Contiguous_sinusoidal_signal(time_sampling), linefmt='r-', markerfmt='ro', basefmt='k-')
    
    ax.axis([0, float(Cycles_of_sinusoid)/f, -1.5, 1.5])
    ax.set_xlabel('Time(s)', fontsize=15)
    ax.set_ylabel('Amplitude', fontsize=15)
    ax.set_title('Sampling Continuous sinusoidal signal', fontsize=20)
```

```python
def graph_signal_reconstructed(ax):
    
    ax.plot(time, Shannon_Function(time, fs))
    ax.stem(time_sampling, Contiguous_sinusoidal_signal(time_sampling), linefmt='r-', markerfmt='ro', basefmt='k-')
    
    ax.axis([-float(Cycles_of_sinusoid)/f, 2*float(Cycles_of_sinusoid)/f, -1.5, 1.5])
    ax.set_xlabel('Time(s)', fontsize=15)
    ax.set_ylabel('Amplitude', fontsize=15)
    ax.set_title('Reconstructed continuous sinusoidal signal', fontsize=20)
```

```python
fig = plt.figure(figsize=(20, 10))
gs1 = gridspec.GridSpec(2, 1)
ax0 = fig.add_subplot(gs1[0])
ax1 = fig.add_subplot(gs1[1])

graph_signal(ax0)
graph_signal_reconstructed(ax1)

plt.show()
```
