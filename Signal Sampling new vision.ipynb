```python
import math
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from numpy import *
```

### Формула Шеннона для реконструкции функции сигнала
Значения f(n/2W) известны и равняются Contiguous_signal_function(ts[n])
<img src="Shannon Function.PNG">
где Xn - значение n-ого семпла

```python
def Shannon_Function(t, signal_sample, fs):
    sum = 0
    #sum1 = 0
    for n in range( len(signal_sample['x']) ):
        sum += signal_sample['y'][n] * np.sinc( 2*fs*t - signal_sample['n'][n] )
    return sum
```

## Функция реконструкции сигнала

```python
# signal_function - исходная функция сигнала, которую будем восстанавливать с помощью теоремы Шеннона
# f - (frequency) часто исходной функции сигнала
# fs - (frequency sampling) частота семплирования исходной функции сигнала
# time_interval - ширина интервала времени сигнала, который будем восстанавливать
def Signal_reconstruction(signal_function, f, fs, time_interval):
    t = linspace(0, time_interval, 10000)
    ts = linspace(-time_interval,
                  2 * time_interval,
                  3 * ( float(time_interval) / ( 1.0/(2*fs) ) ) + 1 )

    signal_sample = {'n' : sort([-i for i in range( int(float(time_interval) / (1.0/(2*fs))) + 1 )] + [i+1 for i in range( 2 * int(float(time_interval) / (1.0/(2*fs))) )]),
                     'x' : ts,
                     'y' : signal_function(ts)}   
    
    # Построение графиков функций сигнала
    fig = plt.figure(figsize=(20, 10))
    gs1 = gridspec.GridSpec(2, 1)
    gs1.update(left=0.1, bottom=0.1, right=0.95, top=0.95, wspace=0.5, hspace=0.3)
    
    # Исходная функция
    ax0 = fig.add_subplot(gs1[0])
    ax0.plot(t, signal_function(t))
    ax0.stem(signal_sample['x'], signal_sample['y'], linefmt='g-', markerfmt='wo', basefmt='k-')
    ax0.axis([-time_interval,
             2 * time_interval,
             min( min(signal_function(ts)), min(signal_function(t)) ),
             max( max(signal_function(ts)), max(signal_function(t)) )])
    ax0.set_xlabel('Time(s)', fontsize=15)
    ax0.set_ylabel('Amplitude', fontsize=15)
    ax0.set_title('Sampling Continuous sinusoidal signal, where f=' + str(f) + ' and fs=' + str(fs), fontsize=20)
    
    # Реконструированная функция
    ax1 = fig.add_subplot(gs1[1])
    ax1.plot(t, Shannon_Function(t, signal_sample, fs))
    ax1.stem(signal_sample['x'], signal_sample['y'], linefmt='g-', markerfmt='wo', basefmt='k-')
    ax1.axis([-time_interval,
             2 * time_interval,
             min( min(signal_function(ts)), min(Shannon_Function(t, signal_sample, fs)) ),
             max( max(signal_function(ts)), max(Shannon_Function(t, signal_sample, fs)) )])
    ax1.set_xlabel('Time(s)', fontsize=15)
    ax1.set_ylabel('Amplitude', fontsize=15)
    ax1.set_title('Reconstructed continuous sinusoidal signal, where f=' + str(f) + ' and fs=' + str(fs), fontsize=20)
    ax1r = ax1.twinx()
    sqe = (signal_function(t) - Shannon_Function(t, signal_sample, fs))**2   # squared error
    ax1r.plot(t, sqe, 'r')
    ax1r.axis([-time_interval, 2*time_interval, -0.2, 0.2])
    ax1r.set_ylabel('squared error', color='r', fontsize=15)
    
    
```

### Исходная функция сигнала

```python
def Contiguous_signal_function_1(t):
    return np.cos( 2 * np.pi * f * t ) + np.sin( np.pi * f * t )
```

```python
f = 10.0  # Hz, signal frequency
fs = 50.0 # Hz, sampling rate (ie. >= 2*f) 
time_interval = 0.35 # Distance from zero
Signal_reconstruction(Contiguous_signal_function_1, f, fs, time_interval)
```
