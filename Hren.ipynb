```python
import math
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from numpy import *
```

```python
fs1 = 30 * 10**3
fs2 = 50 * 10**3
W = 10 * 10**3
```

```python
def s(t):
    return np.cos( 2 * np.pi * W * t )
```

```python
t = linspace(0, float(5)/W, 10000)
y = s(t)
t1 = arange(0, float(5)/W, float(1)/(2*fs1))  ## time index
```

```python
t1
```

```python
plt.plot(t, y)
plt.axis([0, float(5)/W, -1.5, 1.5])
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('Sampling Contiguous sinusoidal signal at fs1=30KHz')
plt.stem(t1, s(t1), linefmt='r-', markerfmt='ro', basefmt='k-')
plt.show()
```

```python
def Shannon_Fun(t):
    summ = 0
    for n in range(1, len(t1)+1, 1):
        summ += s(t1[n-1]) * np.sin( 2*fs1*t*np.pi - n*np.pi ) / ( 2*fs1*t*np.pi - n*np.pi )
    return summ
```

```python
Shannon_Fun(0)
```

```python
Shannon_Fun(1)
```

```python
z = Shannon_Fun(t)
```

```python
plt.plot(t, z)
plt.axis([0, float(5)/W, -1.5, 1.5])
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('Sampling Contiguous sinusoidal signal at fs1=30KHz')
plt.stem(t1, Shannon_Fun(t1), linefmt='r-', markerfmt='ro', basefmt='k-')
plt.show()
```
