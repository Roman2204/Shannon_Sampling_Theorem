```python
import math
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from numpy import *
```

```python
f = 1.0  # Hz, signal frequency
fs = 5.0 # Hz, sampling rate (ie. >= 2*f) 
nCyl=5 #generate five cycles of sinusoid
t = linspace(0, float(nCyl)/f, 1000)
ts = linspace(-float(nCyl)/f,
              2*float(nCyl)/f,
              3 * ( (float(nCyl)/f) / ( float(1)/(2*fs) ) + 1 ) )
Signal_sample = {time : Contiguous_sinusoidal_signal(time) for time in ts}
```

```python
def Contiguous_sinusoidal_signal(t):
    return np.cos( 2 * np.pi * f * t )
```

```python
def graph_signal(ax):
    
    ax.plot(t, Contiguous_sinusoidal_signal(t))
    ax.stem(Signal_sample.keys(), Signal_sample.values(), linefmt='r-', markerfmt='ro', basefmt='k-')
    
    ax.axis([-float(nCyl)/f, 2*float(nCyl)/f, -1.5, 1.5])
    ax.set_xlabel('Time(s)', fontsize=15)
    ax.set_ylabel('Amplitude', fontsize=15)
    ax.set_title('Sampling Continuous sinusoidal signal', fontsize=20)
```

```python
def graph_signal_reconstructed(ax):
    
    ax.plot(time, Shannon_Function(time, fs))
    ax.stem(Signal_sample.keys(), Signal_sample.values(), linefmt='r-', markerfmt='ro', basefmt='k-')
    
    ax.axis([-float(Cycles_of_sinusoid)/f, 2*float(Cycles_of_sinusoid)/f, -1.5, 1.5])
    ax.set_xlabel('Time(s)', fontsize=15)
    ax.set_ylabel('Amplitude', fontsize=15)
    ax.set_title('Reconstructed continuous sinusoidal signal', fontsize=20)
```

```python
fig = plt.figure(figsize=(20, 10))
gs1 = gridspec.GridSpec(2, 1)
ax0 = fig.add_subplot(gs1[0])
ax1 = fig.add_subplot(gs1[1])

graph_signal(ax0)
#graph_signal_reconstructed(ax1)
graph_signal(ax1)

plt.show()
```
